# Load Dataset

```python
import pandas as pd
```

```python
df_barat = pd.read_csv('/content/drive/MyDrive/Pantai/Indonesia Barat/all-task-1.csv')
df_tengah = pd.read_csv('/content/drive/MyDrive/Pantai/Indonesia Tengah/all-task-27.csv')
df_timur = pd.read_csv('/content/drive/MyDrive/Pantai/Indonesia Timur/Pantai wilayah timur.csv')
```

```python
df_all = pd.concat([df_barat, df_tengah, df_timur], ignore_index=True)
```

```python

total_barat = len(df_barat)
total_tengah = len(df_tengah)
total_timur = len(df_timur)
total_all = len(df_all)

print(f"Total data Indonesia Barat: {total_barat}")
print(f"Total data Indonesia Tengah: {total_tengah}")
print(f"Total data Indonesia Timur: {total_timur}")
print(f"Total data keseluruhan: {total_all}")
```

```python
# Drop columns
columns_to_drop = ['is_spending_on_ads', 'competitors', 'website', 'phone', 'can_claim', 'owner', 'main_category', 'categories', 'workday_timing', 'is_temporarily_closed', 'is_permanently_closed', 'closed_on', 'status', 'price_range', 'reviews_per_rating', 'featured_question', 'reviews_link', 'plus_code', 'detailed_address', 'time_zone', 'cid', 'data_id', 'about', 'hours', 'hours', 'most_popular_times', 'popular_times', 'menu', 'reservations', 'order_online_links', 'featured_reviews', 'detailed_reviews', 'query']
df_cleaned = df_all.drop(columns=columns_to_drop, errors='ignore')

df_cleaned.info()
```
