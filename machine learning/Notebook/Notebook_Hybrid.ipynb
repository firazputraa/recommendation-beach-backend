# Prepare

```python
import pandas as pd

# Load dataset
df_place = df
df_review = pd.read_csv('/content/drive/MyDrive/Pantai/Cleaned Dataset/SentimentReview.csv')

# Gabungkan berdasarkan nama tempat
df_merged = df_review.merge(df_place, left_on='place_name', right_on='place_name')

# Buat rating eksplisit dari sentimen
df_merged['sentiment_score'] = df_merged['polarity'].map({'positive': 1, 'neutral': 0, 'negative': -1})
```

```python
df_merged
```

```python
import tensorflow as tf
from sklearn.preprocessing import LabelEncoder

# Encode user_id dan place_id
user_encoder = LabelEncoder()
place_encoder = LabelEncoder()

df_merged['user'] = user_encoder.fit_transform(df_merged['user_id'])
df_merged['place'] = place_encoder.fit_transform(df_merged['place_id'])

num_users = df_merged['user'].nunique()
num_places = df_merged['place'].nunique()

# Dataset tensor
X = df_merged[['user', 'place']]
y = df_merged['sentiment_score'].astype(float)
```

```python
import tensorflow as tf
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Input, Embedding, Flatten, Concatenate, Dense, Dropout, BatchNormalization, LeakyReLU
from tensorflow.keras.regularizers import l2

class HybridRecommender(Model):
    def __init__(self, num_users, num_places, embed_dim=50):
        super(HybridRecommender, self).__init__()

        self.user_embedding = Embedding(
            input_dim=num_users, output_dim=embed_dim, embeddings_regularizer=l2(1e-6)
        )
        self.place_embedding = Embedding(
            input_dim=num_places, output_dim=embed_dim, embeddings_regularizer=l2(1e-6)
        )

        self.dense = tf.keras.Sequential([
            Dense(128),
            LeakyReLU(alpha=0.1),
            BatchNormalization(),
            Dropout(0.5),

            Dense(64),
            LeakyReLU(alpha=0.1),
            BatchNormalization(),
            Dropout(0.3),

            Dense(1)
        ])

    def call(self, inputs):
        user_input, place_input = inputs
        user_vec = Flatten()(self.user_embedding(user_input))
        place_vec = Flatten()(self.place_embedding(place_input))
        x = Concatenate()([user_vec, place_vec])
        return self.dense(x)
```

```python
from tensorflow.keras.callbacks import EarlyStopping

model = HybridRecommender(num_users=num_users, num_places=num_places)
model.compile(optimizer='adam', loss=tf.keras.losses.Huber(), metrics=['mae'])

early_stop = EarlyStopping(
    monitor='val_loss',
    patience=4,
    restore_best_weights=True
)

history = model.fit(
    [df_merged['user'], df_merged['place']],
    y,
    batch_size=256,
    epochs=30,
    validation_split=0.2,
    callbacks=[early_stop],
    verbose=1
)
```

```python
import numpy as np

def recommend_places(user_id_str, top_k=5):
    if user_id_str not in user_encoder.classes_:
        return "User ID tidak ditemukan."

    user_idx = user_encoder.transform([user_id_str])[0]
    all_place_indices = np.arange(num_places)

    user_tensor = tf.convert_to_tensor([user_idx] * num_places)
    place_tensor = tf.convert_to_tensor(all_place_indices)

    predictions = model.predict([user_tensor, place_tensor]).flatten()
    top_k_indices = predictions.argsort()[::-1][:top_k]
    recommended_place_ids = place_encoder.inverse_transform(top_k_indices)

    rekomendasi = df_place[df_place['place_id'].isin(recommended_place_ids)]
    return rekomendasi[['place_id', 'place_name', 'description', 'rating']]
```

```python
def hybrid_search_recommendation(user_id_str, keyword, top_k=5):
    # 1. Validasi user
    if user_id_str not in user_encoder.classes_:
        return "User ID tidak ditemukan di data."

    # 2. Transform user_id ke index internal
    user_idx = user_encoder.transform([user_id_str])[0]

    # 3. Filter tempat yang cocok dengan keyword (di nama atau deskripsi)
    filtered_places = df_place[
        df_place['place_name'].str.contains(keyword, case=False, na=False) |
        df_place['description'].str.contains(keyword, case=False, na=False)
    ]

    if filtered_places.empty:
        return "Tidak ditemukan tempat yang sesuai dengan kata kunci."

    # 4. Transform place_id ke index numerik (yang dikenali oleh model)
    filtered_place_ids = filtered_places['place_id']
    try:
        encoded_place_ids = place_encoder.transform(filtered_place_ids)
    except:
        return "Beberapa tempat tidak dikenali oleh model."

    # 5. Persiapkan input ke model
    user_tensor = tf.convert_to_tensor([user_idx] * len(encoded_place_ids))
    place_tensor = tf.convert_to_tensor(encoded_place_ids)

    # 6. Prediksi skor rekomendasi
    predictions = model.predict([user_tensor, place_tensor], verbose=0).flatten()

    # 7. Urutkan berdasarkan skor tertinggi
    top_indices = predictions.argsort()[::-1][:top_k]
    top_place_ids = filtered_place_ids.iloc[top_indices]

    # 8. Ambil info tempat yang direkomendasikan
    rekomendasi = df_place[df_place['place_id'].isin(top_place_ids)]

    return rekomendasi[['place_id', 'place_name', 'description', 'rating']]
```
