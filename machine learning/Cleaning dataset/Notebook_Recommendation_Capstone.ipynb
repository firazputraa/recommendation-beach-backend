```python
# Drop columns
columns_to_drop = ['is_spending_on_ads', 'competitors', 'website', 'phone', 'can_claim', 'owner', 'main_category', 'categories', 'workday_timing', 'is_temporarily_closed', 'is_permanently_closed', 'closed_on', 'status', 'price_range', 'reviews_per_rating', 'featured_question', 'reviews_link', 'plus_code', 'detailed_address', 'time_zone', 'cid', 'data_id', 'about', 'hours', 'hours', 'most_popular_times', 'popular_times', 'menu', 'reservations', 'order_online_links', 'featured_reviews', 'detailed_reviews', 'query']
df_cleaned = df_all.drop(columns=columns_to_drop, errors='ignore')

df_cleaned.info()
```

```python
# Check for NaN values in the 'description' column
nan_count_rating = df_cleaned['description'].isnull().sum()
print(f"\nNumber of NaN values in 'description' column: {nan_count_rating}")
```

```python
# Check for NaN values in the 'rating' column
nan_count_rating = df_cleaned['rating'].isnull().sum()
print(f"\nNumber of NaN values in 'rating' column: {nan_count_rating}")
```

```python
df_cleaned['place_id'] = ['beach_' + str(i).zfill(4) for i in range(1, len(df_cleaned) + 1)]
```

```python
df_cleaned['description'] = df_cleaned['description'].fillna('-')
```

```python
print(df_cleaned['description'].isnull().sum())
```

```python
# Fungsi untuk menyisakan hanya kata kunci dalam review_keywords
def extract_keywords(review_keywords):
    try:
        # Mengonversi string JSON menjadi list of dictionaries
        review_keywords_list = json.loads(review_keywords)

        # Menyimpan hanya 'keyword' dari setiap dictionary dalam list
        keywords = [item['keyword'] for item in review_keywords_list if 'keyword' in item]

        # Mengembalikan hasil sebagai list kata kunci (tidak dalam format JSON lagi)
        return keywords if keywords else []  # Mengembalikan list kosong jika tidak ada kata kunci yang tersisa
    except (TypeError, json.JSONDecodeError):
        return []  # Jika data tidak valid, kembalikan list kosong

# Memperbarui kolom 'review_keywords' dengan hanya kata kunci
df_cleaned['review_keywords'] = df_cleaned['review_keywords'].apply(extract_keywords)

# Menampilkan hasil untuk memeriksa apakah pemrosesan berjalan dengan benar
print(df_cleaned[['review_keywords']].head())
```

```python
empty_count = df_cleaned[df_cleaned['review_keywords'] == '[]'].shape[0]
print(f"Jumlah baris kosong di kolom 'review_keywords': {empty_count}")
```

```python
df_cleaned = df_cleaned[df_cleaned['review_keywords'].apply(lambda x: len(x) > 0)]
```

```python
df_cleaned['review_keywords'] = df_cleaned['review_keywords'].apply(lambda x: ', '.join(x))
```

```python
import json

def clean_coordinates(coord):
    try:
        # Ubah string JSON ke dict
        coord_dict = json.loads(coord)
        # Ambil latitude dan longitude
        return f"{coord_dict['latitude']}, {coord_dict['longitude']}"
    except (json.JSONDecodeError, TypeError, KeyError):
        return None  # atau '0,0' jika ingin default nilai

# Terapkan ke kolom
df_cleaned['coordinates'] = df_cleaned['coordinates'].apply(clean_coordinates)
```
