```python
timur_len = 1000
barat_len = 1000
tengah_len = len(df_cleaned) - (timur_len + barat_len)

df_cleaned['place_id'] = ''


df_cleaned.iloc[:timur_len, df_cleaned.columns.get_loc('place_id')] = ['timur_' + str(i).zfill(4) for i in range(1, timur_len + 1)]
df_cleaned.iloc[timur_len:timur_len + barat_len, df_cleaned.columns.get_loc('place_id')] = ['barat_' + str(i).zfill(4) for i in range(1, barat_len + 1)]
df_cleaned.iloc[timur_len + barat_len:, df_cleaned.columns.get_loc('place_id')] = ['tengah_' + str(i).zfill(4) for i in range(1, tengah_len + 1)]
```

```python
df_cleaned['description'] = df_cleaned['description'].fillna('-')
```

```python

def extract_keywords(review_keywords):
    try:

        review_keywords_list = json.loads(review_keywords)

        keywords = [item['keyword'] for item in review_keywords_list if 'keyword' in item]


        return keywords if keywords else []
    except (TypeError, json.JSONDecodeError):
        return []  #


df_cleaned['review_keywords'] = df_cleaned['review_keywords'].apply(extract_keywords)

print(df_cleaned[['review_keywords']].head())
```

```python
empty_count = df_cleaned[df_cleaned['review_keywords'] == '[]'].shape[0]
print(f"Jumlah baris kosong di kolom 'review_keywords': {empty_count}")
```

```python
df_cleaned = df_cleaned[df_cleaned['review_keywords'].apply(lambda x: len(x) > 0)]
```

```python
df_cleaned['review_keywords'] = df_cleaned['review_keywords'].apply(lambda x: ', '.join(x))
```

```python
print(df_cleaned['review_keywords'])
```

```python
def clean_coordinates(coord):
    try:

        coord_dict = json.loads(coord)

        return f"{coord_dict['latitude']}, {coord_dict['longitude']}"
    except (json.JSONDecodeError, TypeError, KeyError):
        return None


df_cleaned['coordinates'] = df_cleaned['coordinates'].apply(clean_coordinates)
```

```python
df_cleaned = df_cleaned.drop(columns=['images'])
```

```python


def clean_address(address):
    address = re.sub(r'\b[0-9A-Za-z]+[\+][0-9A-Za-z]+\b', '', address)

    address = re.sub(r'\d{5}$', '', address)

    address = re.sub(r'\s+', ' ', address).strip()
    return address


df_cleaned['address'] = df_cleaned['address'].astype(str).fillna('')


df_cleaned['address'] = df_cleaned['address'].apply(clean_address)

print(df_cleaned[['address']].head())
```

```python
def clean_address(address):

    address = address.strip(',')
    address = re.sub(r'([A-Za-z])([A-Z])', r'\1 \2', address)


    address = re.sub(r'\s*,\s*', ', ', address)

    address = re.sub(r',\s*,', ',', address)


    address = re.sub(r',\s+', ', ', address)

    return address.strip()

df_cleaned['address'] = df_cleaned['address'].apply(clean_address)

print(df_cleaned[['address']].head())
```
